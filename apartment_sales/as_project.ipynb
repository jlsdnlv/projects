## Предобработка данных

В этом разделе необходимо изучить пропуски, рассмотреть типы данных столбцов, устранить дубликаты и выбивающиеся значения.

### Обработка пропусков

```python
# Выведем количество пропусков по столбцам
df.isna().sum()
```

Из 22 представленных столбцов только в 8 отсутствуют пропущенные значения.

**Площадь кухни `kitchen_area`**

Если квартира относится к студии или свободной планировке, то в столбец `kitchen_area` необходимо поставить 0.


```python
# Если квартира - студия
df['kitchen_area'] = df['kitchen_area'].where(~((df['kitchen_area'].isna()) & (df['studio'])), 0)
```

```python
# Если квартира - свободной планировки
df['kitchen_area'] = df['kitchen_area'].where(~((df['kitchen_area'].isna()) & (df['open_plan'])), 0)
```

Чтобы заполнить оставшиеся пропуски, сначала найдем соотношение медианной площади кухни к медианной общей. Затем заполним пропуски в столбце `kitchen_area` умножением общей площади на найденное соотношение.

```python
total_area_kitchen_area_ratio = df['kitchen_area'].median() / df['total_area'].median()
```

```python
df['kitchen_area'] = df['kitchen_area'].fillna(df['total_area'] * total_area_kitchen_area_ratio)
```

**Площадь жилой зоны `living_area`:** 

Сделаем то же самое как и в столбце `kitchen_area` для заполнения пропусков.

```python
total_area_living_area_ratio = df['living_area'].median() / df['total_area'].median()
```

```python
df['living_area'] = df['living_area'].fillna(df['total_area'] * total_area_living_area_ratio)
```

**Число балконов `balcony`:**

Если продавец не указал число балконов, то, скорее всего, в его квартире их нет. Такие пропуски правильно заменить на 0.

```python
df['balcony'] = df['balcony'].fillna(0)
```

**Высота потолков `ceiling_height`**:

Пропуски в этом столбце можно заменить медианным значением.

```python
# Выведем медианное значение высоты потолка
df['ceiling_height'].median()
```

Медианное значение высоты потолков составляет 2.65 м, что похоже на правду.

```python
df['ceiling_height'] = df['ceiling_height'].where(df['ceiling_height'].notna(), df['ceiling_height'].median())
```

Еще раз проверим количество пропусков по столбцам:

```python
df.isna().sum()
```

По оставшимся столбцам нет возможности предложить логичную замену пропускам:
   - картографические данные нецелесообразно заменять медианными значениями. Размах выборки большой, поэтому заполнение пропусков в таких столбцах вероятнее всего приведет к неактуальности данных.
   - столбец `days_exposition` невозможно определить. Из известного у нас есть только дата публикации объявления.
   - остальные столбцы `floors_total`, `is_apartment` также не получилось заменить, но они и не сильно понадобятся в дальнейшем.

### Устранение дубликатов

Явных дубликатов в таблице нет. Неявные дубликаты могут встречаться только в столбце `locality_name`, так как во всех остальных столбцах численные показатели.

```python
# Выведем все уникальные значения
df['locality_name'].sort_values().unique()
```

```python
# Выведем количество уникальных значений
df['locality_name'].nunique()
```

Можно увидеть, что в некоторых словах используется буква `ё`. Избавимся от них путем замены на букву `е`.

```python
df['locality_name'].replace('ё', 'е', inplace=True, regex=True)
```

Также в столбце `locality_name` есть одни и те же населенные пункты написанные по-разному. Например, «поселок Рябово» и «поселок городского типа Рябово». Избавимся от таких неявных дубликатов с помощью регулярных выражений.

```python
to_replace = {
     r'(\b)поселок городского типа(\b)':r'\1поселок\2',
     r'(\b)коттеджный поселок(\b)':r'\1поселок\2',
     r'(\b)городской поселок(\b)':r'\1поселок\2',
     r'(\b)поселок при железнодорожной станции(\b)':r'\1поселок\2',
     r'(\b)поселок станции(\b)':r'\1поселок\2',
     r'(\b)садоводческое некоммерческое товарищество(\b)':r'\1садовое товарищество\2'
}

df['locality_name'] = df['locality_name'].replace(to_replace, regex=True)
```

```python
# Выведем новое количество уникальных значений
df['locality_name'].nunique()
```

Таким образом, снизили количество уникальных значений с 364 до 318. Пропуски в этом столбце нет возможности заполнить.

### Изменение типов данных

Столбцы `last_price`, `floors_total`, `balcony`, `airports_nearest`, `cityCenters_nearest`, `parks_around3000`, `parks_nearest`, `ponds_around3000`, `ponds_nearest`, `days_exposition` определены как тип `float`. Дробные значения в этих столбцах не имеют смысла, поэтому будет целесообразнее перевести их в тип `int`.
```python
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

Выведем общую информацию по таблице, чтобы проверить изменились ли типы данных.

```python
df.info()
```

У изменения типов данных есть преимущества:
- Если значения в столбце - целые числа, использование типа `int` обеспечит более точное и четкое представление данных. 
- Тип данных `int` занимает меньше памяти, чем `float`, что важно, если объем данных большой.
- Чтобы проводить с датой арифметические операции, необходимо дату переводить в особый тип данных - `datetime`.

### Устранение выбивающихся значений

**Столбец `ceiling_height`**

Примем, что высота потолков должна находится в диапазоне от 2 до 5 м.

В столбце `ceiling_height` может быть указана высота потолков 25 м и 32 м. Логично предположить, что на самом деле это вещественные значения: 2.5 м и 3.2 м. Таким образом, значения от 20 до 50 м нужно поделить на 10.

```python
df['ceiling_height'] = df['ceiling_height'].apply(lambda height: height/10 if 20 <= height <= 50 else height)
```

Всё остальное - выбивающиеся значения, которые надо удалить.

```python
df = df[(df['ceiling_height'] >= 2) & (df['ceiling_height'] <= 5)]
```

## Добавление новых столбцов

```python
# сбросим индексы, чтобы без проблем создавать новые столбцы
df = df.reset_index(drop=True)
```

**Цена одного квадратного метра**

```python
df['square_meter_price'] = (df['last_price'] / df['total_area']).astype('int')
```

**День недели публикации объявления**

```python
# 0 — понедельник, 1 — вторник и так далее
df['publication_weekday'] = df['first_day_exposition'].dt.weekday
```

**Месяц публикации объявления**

```python
df['publication_month'] = df['first_day_exposition'].dt.month
```

**Год публикации объявления**

```python
df['publication_year'] = df['first_day_exposition'].dt.year
```

**Тип этажа квартиры**

Значения - "первый", "последний", "другой".

```python
# Напишем функцию, которая будет определять тип этажа квартиры
def define_floor_type(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    return 'другой'
```

```python
df['floor_type'] = df.apply(define_floor_type, axis=1)
```

**Расстояние до центра города в километрах**

```python
df['cityCenters_nearest_in_km'] = (df['cityCenters_nearest'] / 1000).round(0)
```

### Какие факторы больше всего влияют на общую стоимость объекта?

Чтобы оценить влияние различных факторов на стоимость объекта будем строить диаграммы рассеяния и посчитаем коэффициент корреляции. Будем использовать сводные таблицы для подготовки данных перед визуализацией.

**Общая площадь**

```python
# Сводная таблица
pivot_table_total_area = df.pivot_table(index='total_area', values='last_price', aggfunc='median')
# Преобразуем индекс обратно в столбец
pivot_table_total_area = pivot_table_total_area.reset_index()
# Строим диаграмму рассеяния
plt.figure(figsize=(8, 6))
sns.scatterplot(data=pivot_table_total_area, y='last_price', x='total_area')
plt.ylabel('Цена')
plt.xlabel('Общая площадь')
plt.title('Диаграмма рассеяния между ценой и общей площадью квартиры', fontsize=10)
plt.show()
```

```python
# Коэффициент корреляции
print(f"Коэффициент корреляции: {round(df['total_area'].corr(df['last_price']), 2)}")
```

Исходя из полученного графика и коэффициента корреляции, мы можем сказать, что между общей площадью и ценой квартиры существует довольно сильная положительная зависимость. Иными словами, при увеличении общей площади квартиры вероятно увеличится и её стоимость, и наоборот.

Положительная корреляция в данном контексте логична: чем больше общая площадь квартиры, тем дороже квартира.

**Жилая площадь**

```python
# Сводная таблица
pivot_table_living_area = df.pivot_table(index='living_area', values='last_price', aggfunc='median')
# Преобразуем индекс обратно в столбец
pivot_table_living_area = pivot_table_living_area.reset_index()
# Строим диаграмму рассеяния
plt.figure(figsize=(8, 6))
sns.scatterplot(data=pivot_table_living_area, y='last_price', x='living_area')
plt.ylabel('Цена')
plt.xlabel('Жилая площадь')
plt.title('Диаграмма рассеяния между ценой и жилой площадью квартиры', fontsize=10)
plt.show()
```

```python
# Коэффициент корреляции
print(f"Коэффициент корреляции: {round(df['living_area'].corr(df['last_price']), 2)}")
```

Аналогичная ситуация как с общей площадью квартиры, но коэффициент корреляции слегка меньше. Это означает, что увеличение жилой площади часто сопровождается увеличением стоимости квартиры, и наоборот.
